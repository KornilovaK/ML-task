```python
import numpy as np
import pandas as pd
from transformers import pipeline
```

```python
classifier = pipeline("zero-shot-classification")
df = pd.read_csv('cleared_df.csv')
df.isna().sum()
```

```python
df = df.dropna()
texts = df.text.values
classes = ['ham', 'spam']
```

```python
preds = []
for i in range(len(texts)):
  pred = classifier(texts[i], candidate_labels=classes)
  preds.append(np.argmax(pred['scores']))
  if i % 100 == 0:
    print(i)
```

долго обрабатывается, поэтому спустя некоторое время уже можно останавливать, чтобы посмотреть на промежуточные результаты

```python
length = len(preds)
y = df.text_type.values[:length]
len(y) == len(preds)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y, preds)
```

0.5 Что означает абсолютно рандомные, неправильные предсказания
